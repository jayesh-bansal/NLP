```python
#IMPORTING THE NECESSARY LIBRARYS.
import nltk
from nltk.tokenize import word_tokenize,sent_tokenize
nltk.download('punkt')
import pandas as pd
```

```python
#DATA TO BE TESTED
data = {
    'Text':[
        "Hello my name is DEEPAK RAMGIRI.",
        "I am learning NATURAL LANGUAGE PROGRAMMING.",
        "Python is a programming language. And it is used widely to handle nlp operations.",
        "This is example of segmentation.",
        "NLTK is a useful library for text processing.",
        "that's it now",
    ]
}

df = pd.DataFrame(data)
df.head()
#PRINTING THE FIRST 10 VALUES FROM THE DATA
```

```python
#SENTENCE TOKENIZATION
df['Sentences'] = df['Text'].apply(sent_tokenize)
df[['Text','Sentences']].head()
#PRINTING THE FIRST 10 VALUES OF THE TEXT DATA AND SENTENCES HEAD
```

```python
#APPLYING WORD TOKENZATION  TO THE TEXT FILE
df['Words'] = df['Text'].apply(word_tokenize)

#PRINTING THE HEAD VALUES
df[['Text', 'Words']].head()
```

```python
# TO DOWNLOAD THE DATA OR FILE
segmented_file_path = "segemented_text_dataset.csv"
df.to_csv(segmented_file_path,index="False")
from google.colab import files
files.download(segmented_file_path)
#A FILE DOWNLOADED AFTER THIS
```

# STEMMING

```python
from nltk.stem import PorterStemmer
```

```python

data = {
    'Text': [
        "I'am  a Student at Sr university.",
        "I'am  currently a  B.tech 3rd Year Student.",
        "So these is a nlp lab and iam doning coding to learn nlp.",
        "and im hungry even though i just ate my breakfast"
    ]
}
df = pd.DataFrame(data)
df.head()
```

```python
ps = PorterStemmer()

#Applying the Stem words to the given data.
def stem_words(text):
  words = word_tokenize(text)
  stemmed_words = [ps.stem(word) for word in words]
  return ''.join(stemmed_words)
```

```python
#Applying the Stem Words
df['Stemmed_Text'] = df['Text'].apply(stem_words)
df[['Text','Stemmed_Text']].head()
#Printing the head values
```

```python
stemmed_file_path = 'stemmed_text_dataset.csv'
df.to_csv(stemmed_file_path,index=False)
from google.colab import files
files.download(stemmed_file_path)
#Downloading the file as a CSV File
```
